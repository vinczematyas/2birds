```python
import torch
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from dataclasses import dataclass

from utils import SignalDataset
```

```python
@dataclass
class Args():
    batch_size: int = 5

args = Args()
```

```python
train_ds = SignalDataset(torch.load('train_dataset.pt'))
train_dl = DataLoader(train_ds, batch_size=args.batch_size)
```

```python
x,y = next(iter(train_dl))
```

```python
for idx in range(args.batch_size):
    plt.plot(x[idx].squeeze())
    plt.plot(y[idx].squeeze())
    plt.show()
```

---

```python
sub_ds = SignalDataset(torch.load('result_dataset.pt'))
sub_dl = DataLoader(sub_ds, batch_size=args.batch_size)
```

```python
x_sub, y_sub = next(iter(sub_dl))
```

```python
for idx in range(args.batch_size):
    plt.plot(x_sub[idx].squeeze())
    plt.plot(y_sub[idx].squeeze())
    plt.show()
```
